```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split

df = pd.read_csv("epl_player_stats_24_25.csv")
df = df.drop('Conversion %', axis=1)
df = df.drop('Shots On Target', axis=1)
df.columns = df.columns.str.replace(' ', '', regex=False)
df.columns = df.columns.str.strip()
df['PlayerName'] = df['PlayerName'].str.upper()
df['Club'] = df['Club'].str.upper()
df['Nationality'] = df['Nationality'].str.upper()
df['Position'] = df['Position'].str.upper()
df['Saves%'] = df['Saves%'].str.rstrip('%').astype(float) / 100
df['Passes%'] = df['Passes%'].str.rstrip('%').astype(float) / 100
df['Crosses%'] = df['Crosses%'].str.rstrip('%').astype(float) / 100
df['fThirdPasses%'] = df['fThirdPasses%'].str.rstrip('%').astype(float) / 100
df['gDuels%'] = df['gDuels%'].str.rstrip('%').astype(float) / 100
df['aDuels%'] = df['aDuels%'].str.rstrip('%').astype(float) / 100
df.loc[df['PlayerName'] == 'EMERSON', 'Position'] = 'DEF'
df.loc[df['PlayerName'] == 'JOSH KING', 'Position'] = 'MID'
df.loc[df['PlayerName'] == 'ANTOINE SEMENYO', 'Goals'] = 11
df.loc[df['PlayerName'] == 'EVANILSON', 'Goals'] = 10
df.loc[df['PlayerName'] == 'RAÚL JIMÉNEZ', 'Goals'] = 12
df.loc[df['PlayerName'] == 'LIAM DELAP', 'Goals'] = 8
df.loc[df['PlayerName'] == 'JØRGEN STRAND LARSEN', 'Goals'] = 14
df.loc[df['PlayerName'] == 'JUSTIN KLUIVERT', 'Goals'] = 12
df.loc[df['PlayerName'] == 'KEVIN SCHADE', 'Goals'] = 11
df.loc[df['PlayerName'] == 'LUIS DÍAZ', 'Goals'] = 13
df.loc[df['PlayerName'] == 'BRENNAN JOHNSON', 'Goals'] = 11
df.loc[df['PlayerName'] == 'JARROD BOWEN', 'Goals'] = 13
```

```python
gkps_df = df[df['Position'] == 'GKP']
defs_df = df[df['Position'] == 'DEF']
mids_df = df[df['Position'] == 'MID']
fwds_df = df[df['Position'] == 'FWD']
```

```python
train_cols = ["Minutes","Shots","BigChancesMissed", "HitWoodwork", "Offsides","Touches"]
X = torch.tensor(df[train_cols].values, dtype=torch.float)
y = torch.tensor(df['Goals'].values, dtype=torch.float).view(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)
```

```python
torch.manual_seed(42)
model = nn.Sequential(
    nn.Linear(6, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1)
)
```

```python
criteron = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 10000
for epoch in range(num_epochs):
    predictions = model(X)
    MSE = criteron(predictions,y) 
    optimizer.zero_grad()
    MSE.backward() 
    optimizer.step() 
```

```python
torch.save(model, 'model.pth')

# Loading PyTorch models
loaded_model = torch.load('model.pth', weights_only=False)
```

```python
loaded_model.eval()
with torch.no_grad():
    test_predictions = loaded_model(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Convert predictions to NumPy array if needed
y_true = y_test.cpu().numpy()
y_pred = test_predictions.detach().cpu().numpy().squeeze()

mse = mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f"Test MSE: {mse:.3f}")
print(f"Test MAE: {mae:.3f}")
print(f"Test R²: {r2:.3f}")
```
